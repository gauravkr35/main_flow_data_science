```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
dataset=pd.read_csv("USvideos.csv")
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.duplicated().sum()
```

```python
dataset.drop_duplicates(inplace=True)
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
dataset.drop(columns=['thumbnail_link','description'] , axis=1,inplace=True)
```

```python
dataset
```

```python
dataset['trending_date']=dataset['trending_date'].apply(lambda x: datetime.strptime(x, '%y.%d.%m'))
```

```python
dataset.head()
```

```python
dataset['publish_time']=pd.to_datetime(dataset['publish_time'])
dataset.head()
```

```python
dataset['publish_month']=dataset['publish_time'].dt.month
dataset['publish_day']=dataset['publish_time'].dt.day
dataset['publish_hour']=dataset['publish_time'].dt.hour
dataset.head()
```

```python
sorted(dataset['category_id'].unique())
```

```python
dataset['year']=dataset['publish_time'].dt.year
```

```python
dataset
```

```python
yearly_count=dataset.groupby('year')['video_id'].count()
```

```python
yearly_count
```

```python
yearly_count.plot(kind='bar',xlabel='year',ylabel='Total Publish count',title='Total Publish video per year')
plt.show()
```

```python
yearly_views=dataset.groupby('year')['views'].sum()
```

```python
yearly_views
```

```python
yearly_views.plot(kind='bar',xlabel='year',ylabel='Total Views',title='Total views per year')
plt.show()
```

```python
video_per_hour=dataset['publish_hour'].value_counts().sort_index()
```

```python
video_per_hour
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=video_per_hour.index , y=video_per_hour.values)
plt.title('Number of videos published per Hour')
plt.xlabel('Hour of day')
plt.ylabel('Number of video')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='views',y='likes',data=dataset)
plt.title('Views VS Likes')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()
```
