```python
for i in [1,9,12,16,19,21]:   
    Q1=np.percentile(dataset[dataset.columns[i]],25)
    Q3=np.percentile(dataset[dataset.columns[i]],75)
    IQR=Q3-Q1
    min_range=Q1-1.5*IQR
    max_range=Q3+1.5*IQR
    dataset=dataset[dataset[dataset.columns[i]]>min_range]
    dataset=dataset[dataset[dataset.columns[i]]<max_range]
```
