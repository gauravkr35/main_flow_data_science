```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("tips_new.csv")
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
plt.scatter(x="day" , y="tip" , data=dataset)
plt.title("Scatter Plot")
plt.xlabel("Day")
plt.ylabel("Tip")
plt.show()
```

```python
plt.scatter(x="day" , y="tip" , data=dataset)
plt.title("Scatter Plot")
plt.xlabel("Day")
plt.ylabel("Tip")
plt.colorbar()
plt.show()
```

```python
plt.plot(dataset['tip'])
plt.plot(dataset['size'])
plt.title("Scatter Plot")
plt.xlabel("Day")
plt.ylabel("Tip")
plt.show()
```

```python
plt.hist(x="day"  , data=dataset  )
plt.title("Histogram")
plt.xlabel("Day")
plt.ylabel("Tip")
plt.show()
```
